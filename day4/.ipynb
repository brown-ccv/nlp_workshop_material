### Design a function that writes the top n frequently occurring words in a corpus to a CSV file. The function should have two arguments: corpus and n. You can test your function on the abcnews data provided below.

```python
import pandas as pd
df = pd.read_csv('abcnews-date-text.csv')
print(df.head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def get_top_n_words(corpus, n=10):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    freq_words = pd.DataFrame(words_freq, columns = ['word', 'frequency'])
    freq_words.sort_values(by='frequency', ascending = False, inplace=True)
    n_freq_words = freq_words[:n]
    n_freq_words.to_csv('./{}_most_frequent_words.csv'.format(n), index=False)
    
    return n_freq_words
```

```python
corpus = df['headline_text']
get_top_n_words(corpus, 20)
```
